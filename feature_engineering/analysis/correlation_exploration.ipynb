```python
import pandas as pd
```

```python
df = pd.read_csv('../annotated.csv', delimiter=';')
```

```python
df
```

```python
df = df.drop(columns=['document_id','range_from','range_to'])
```

```python
df.corr()
```

```python
corr = df.corr()
```

```python
import matplotlib
import seaborn as sns
svm = sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```

```python
figure = svm.get_figure()    

figure.savefig('svm_conf.png', dpi=250, bbox_inches = "tight")
```
