# NER Data analysis

```python
import pandas as pd
```

```python
df = pd.read_csv('annotated.csv')
```

```python
def count_values(df, column, min_value, xlabel):
    ner_list = []
    for index, row in df.iterrows():
        val = row[column]
        if type(val) == str:
            ner_object = row[column].split(', ')
            for ner in ner_object:
                ner_list.append(ner)
    df_ner = pd.DataFrame(ner_list)
    df_ner = df_ner[df_ner[0] != 'his']
    df_ner = df_ner[df_ner[0] != 'He']
    df_ner = df_ner[df_ner[0] != 'he']
    df_ner = df_ner[df_ner[0] != 'her']
    df_ner = df_ner[df_ner[0] != 'she']
    df_ner = df_ner[df_ner[0] != 'him']
    df_ner = df_ner[df_ner[0] != 'She']
    df_ner = df_ner[0].value_counts()
    df_ner = df_ner[df_ner > min_value]
    plot = df_ner.plot.bar()
    plot.set_xlabel(xlabel)
    plot.set_ylabel("Occurrence")
    plot.get_figure().savefig('plot.pdf', dpi=200, bbox_inches = "tight")
```

```python
count_values(df, 'ner_person', 11, 'Person')
```

```python
count_values(df, 'ner_organization', 5, "Organization")
```

```python
count_values(df, 'ner_country', 10, "Country")
```
