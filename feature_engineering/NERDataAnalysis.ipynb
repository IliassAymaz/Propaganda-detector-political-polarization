# NER Data analysis

```python
import pandas as pd
```

```python
df = pd.read_csv('annotated.csv')
```

```python
def count_values(df, column, min_value):
    ner_list = []
    for index, row in df.iterrows():
        val = row[column]
        if type(val) == str:
            ner_object = row[column].split(', ')
            for ner in ner_object:
                ner_list.append(ner)
    df_ner = pd.DataFrame(ner_list)
    df_ner = df_ner[0].value_counts()
    df_ner = df_ner[df_ner > min_value]
    df_ner.plot.bar()
```

```python
count_values(df, 'ner_person', 10)
```

```python
count_values(df, 'ner_organization', 4)
```

```python
count_values(df, 'ner_country', 4)
```
